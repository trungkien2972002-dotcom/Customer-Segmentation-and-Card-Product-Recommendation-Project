1. Nhóm biến phân phối chữ U:PURCHASES_INSTALLMENTS_FREQUENCY và PURCHASE_FREQUENCY
- Biểu đồ chữ U phản ánh sự phân cực rõ rệt trong thói quen của chủ thẻ:
    + Cột 0: Nhóm khách hàng hoàn toàn ngủ đông, hoặc mở thẻ ra nhưng không bao giờ quẹt thẻ (hoặc không bao giờ mua trả góp)
    + Cột 1: Nhóm khách hàng coi thẻ tín dụng là vật bất ly thân, quẹt thẻ liên tục trong mọi giao dịch phát sinh hàng tháng.
- Phương án xử lý: Giữ nguyên (không biến đổi log). Vì giá trị của 2 biến này đã bị giới hạn trong khoảng từ 0 đến 1, chúng không tạo ra các outlier làm méo khoảng cách hình học. Định dạng chữ U này là trọng tâm giúp K-means bóc tách nhóm khách hàng tích cực và tiêu cực.
2. Nhóm biến lệch phải (BALANCE_FREQUENCY và TENURE)
- BALANCE_FREQUENCY lệch phải nặng (tập trung sát vạch 1): Chứng tỏ đại đa số khách hàng (>80%) có số dư tài khoản được cập nhật liên tục hằng tháng. Họ dùng thẻ rất đều đặn, số người mở thẻ rồi vứt xó (gần vạch 0) là cực kỳ ít.
- TENURE lệch phải (tập trung tại ô số 12): Phản ánh phần lớn tệp chủ thẻ trong dữ liệu là khách hàng trung thành, đã gắn bó với ngân hàng trọn vẹn cả 12 tháng. Số lượng thẻ mới mở (6-11 tháng) chiếm tỷ trọng nhỏ.
- Phương án xử lý: GIỮ NGUYÊN (Không biến đổi Log). Hàm Log chỉ có tác dụng nén đuôi dài bên phải. Với dữ liệu lệch trái và có khoảng giá trị nhỏ/rời rạc (như TENURE từ 6 đến 12), không cần biến đổi toán học gì thêm.
3. Nhóm biến số lệch trái nặng (Các biến còn lại)
- Phản ánh quy luật Pareto (80/20) kinh điển trong ngành tài chính. Phần lớn mọi người chi tiêu, rút tiền ở mức trung bình và thấp (tập trung quanh vạch 0), nhưng doanh số và lợi nhuận của ngân hàng lại bị kéo dài bởi một nhóm nhỏ các "cá mập" chi tiêu hàng chục nghìn USD ở đuôi bên phải.
- Phương án xử lý: BẮT BUỘC áp dụng Log Transformation (np.log1p). Bước này sẽ kéo các giá trị cực lớn ở đuôi phải về gần trung tâm hơn, biến phân phối thô dốc đứng thành phân phối tiệm cận hình chuông chuẩn.

```python
df_cleaned.drop(['CUST_ID'], axis=1, inplace=True)
```

```python

# Danh sach cac bien giu nguyen
columns_to_keep = ['BALANE_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY', 'PURCHARES_FREQUENCY', 'TENURE']

# Danh sach cac bien can bien doi (Lech trai)
columns_to_log = [col for col in df_cleaned.columns if col not in columns_to_keep]

# Tien hanh xu ly du lieu phan tach
df_transformed = df_cleaned.copy()

# Ap dung log1p cho nhom lech trai
df_transformed[columns_to_log] = df_transformed[columns_to_log].apply(np.log1p)

# Chuan hoa toan bo tap du lieu ve mean=0, Std=1 (bat buoc cho K-means)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_transformed)

# Chuyen hoa thanh dataframe hoan chinh de nap vao thuan toan
df_final_scaled = pd.DataFrame(X_scaled, columns=df_cleaned.columns)

print(df_final_scaled.shape)
```

Phương pháp đường cong khuỷu tay (Elbow Method)
Phương pháp này tính toán tổng bình phương khoảng cách từ các điểm dữ liệu đến tâm cụm của chúng (gọi là chỉ số WCSS hoặc Inertia). Khi số lượng cụm tăng lên, WCSS sẽ giảm dần. Bạn cần tìm điểm mà tại đó tốc độ giảm của WCSS bắt đầu chậm lại rõ rệt (tạo thành một cái "khuỷu tay").

```python
# Ve bieu do Elbow
from sklearn.cluster import KMeans

# Khoi tao danh sach de luu gia tri WCSS (Inertia)
wcss = []

# Chay thu thuat toan voi so cum k thay doi tu 1 den 10
# Su dung bien du lieu da chuan hoa
for k in range(1,11):
    # random_state = 42 de dam bao ket quan khong thay doi sau moi lan chay
    kmeans = KMeans(n_clusters=k, init = 'k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Ve bieu do Elbow
plt.figure(figsize=(10,6))
plt.plot(range(1,11), wcss, marker='o', linestyle='--', color='teal')
plt.title('Elbow Method', fontsize = 14)
plt.xlabel('K', fontsize = 12)
plt.ylabel('WCSS (Inertia)', fontsize = 12)
plt.xticks(range(1,11))
plt.grid(True, alpha = 0.3)
plt.show()
```

Đôi khi biểu đồ Elbow không cho một cái "khuỷu tay" rõ ràng (đường cong bị mượt quá). Lúc này, DA sẽ dùng thêm Silhouette Score để đo độ tự tin của thuật toán. Điểm số này chạy từ -1 đến 1:
- Càng gần 1: Các cụm phân tách nhau rất rõ ràng, không bị chồng lấn.
- Càng gần 0: Các cụm nằm quá sát hoặc dính vào nhau.

```python
# In ra diem so cu the cho tung K (Kiem tra tu 2 den 6 cum)

from sklearn.metrics import silhouette_score

# Kiem tra Silhouette_score tu 2 den 6 cum
for k in range(2,7):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    cluster_labels = kmeans.fit_predict(X_scaled)

    # Tinh diem trung binh cua silhouette_score 
    silhouette_avg = silhouette_score(X_scaled, cluster_labels)
    print(f'With K = {k}, , silhouette_avg = {silhouette_avg:.4f}')
```

Điểm bẻ gãy (Khuỷu tay): Khi quan sát kỹ đường cong WCSS, tốc độ giảm (độ dốc) bắt đầu thay đổi rõ rệt nhất tại các điểm từ $K = 3$ sang $K = 4$. Sau $K = 4$, đường cong bắt đầu thoải dần và đi xuống một cách tuyến tính đều đặn (từ 4 lên 5, 5 lên 6 độ dốc không còn thay đổi đột ngột nữa). Do đó, vùng giao thoa toán học lý tưởng nằm ở $K = 3$ hoặc $K = 4$.

- Thực tế dữ liệu: Điểm Silhouette ở mức 0.2 cho thấy các cụm có ranh giới đan xen lẫn nhau chứ không tách biệt hoàn toàn thành các khối cô lập trong không gian đa chiều.
- Lý do: Đây là đặc tính hoàn toàn bình thường của dữ liệu hành vi tài chính/thẻ tín dụng. Hành vi tiêu dùng của con người là một dải liên tục (ví dụ: một người thích trả góp vẫn có lúc rút tiền mặt hoặc quẹt thẻ thông thường). Dữ liệu thực tế (Real-world data) hiếm khi cho điểm Silhouette đẹp như sách giáo khoa ($>0.5$).
- Lựa chọn: Thường với điểm số thấp đều như vậy, ta sẽ ưu tiên chọn số cụm mang lại ý nghĩa nghiệp vụ rõ ràng nhất thay vì quá phụ thuộc vào sự chênh lệch nhỏ của điểm số. Với K = 4 sẽ khớp hoàn hảo với chiến lược sản phẩm thẻ mà BA đã đề xuất trong BRD trước đó

-> Bước tiếp theo, gán nhãn cụm:
- Cụm 1: Khách hàng tiêu dùng cao cấp, trả thẳng (Transactors).
- Cụm 2: Khách hàng thích nợ xoay vòng, mua trả góp (Revolvers / Installment Users).
- Cụm 3: Khách hàng lạm dụng rút tiền mặt rủi ro cao (Cash Borrowers).
- Cụm 4: Khách hàng mở thẻ rồi để đó, chi tiêu cực thấp (Inactive Users). 

```python
# Khoi tao va huan luyen mo hinh K-means voi K = 4
kmeans_final = KMeans(n_clusters=4, init='k-means++', random_state=42)

# X_scaled la mang du lieu da qua xu ly StandardScaler
cluster_labels = kmeans_final.fit_predict(X_scaled)

# Gan nhan cum nguoc lai vao Dataframe ban dau
df_cleaned['CLUSTER'] = cluster_labels
print(df_cleaned['CLUSTER'].value_counts(normalize=True).round(3)*100)
```
